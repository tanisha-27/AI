```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class DataRead():
    def __init__(self,fname):
        self.data = pd.read_csv(fname)
        
    def RowSelect(self,x,y):
        print(self.data.iloc[x-1,y-1])
        print(self.data.iloc[y-1])
        
        
class DataRead2(DataRead):
    def CorrelationPlot(self):
        data1 = self.data
        correlation_plot = data1.corr()
        sns.heatmap(correlation_plot, annot = True)
        plt.show()
    
    def RowSelect(self,x,y,z):
        print(self.data.iloc[x-1])
        print(self.data.iloc[y-1])
        print(self.data.iloc[z-1])
```

```python
a = DataRead2("C:/Users/User/Downloads/sales_data_sample.csv")
a.CorrelationPlot()
```

```python
a.RowSelect(1,2,3)
```

```python
filepath = input("Enter the file path\n")
option = input ("Enter 1: Correlation Plot \nEnter 2: Display Rows\n")
option = int(option)
if (option == 1):
    df = DataRead2(filepath)
    df.CorrelationPlot()
elif (option == 2):
    x = input("Enter Row:\n")
    x = int(x)
    y = input("Enter Row:\n")
    y = int(y)
    z = input("Enter Row:\n")
    z = int(z)
    df = DataRead2(filepath)
    df.RowSelect(x,y,z)
else:
    print("Invalid option")
```
